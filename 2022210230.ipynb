```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 加载数据集
df = pd.read_csv('data/Customer Churn.csv')
```

```python
# 显示数据集的基本信息
print("数据集信息：")
print(df.info())
```

```python
# 显示数据集的前几行
print("\n数据集的前几行：")
print(df.head())
```

```python
# 检查数据集中是否有缺失值
print("\n数据集中的缺失值：")
print(df.isnull().sum())
```

```python
# 描述数据集的统计信息
print("\n数据集的统计摘要：")
print(df.describe())
```

```python
# 可视化每个属性的分布情况
df.hist(bins=30, figsize=(20, 15))
plt.show()
```

```python
# 可视化属性之间的相关性矩阵
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# 可视化目标变量'Churn'的分布情况
plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=df)
plt.title('Churn的分布情况')
plt.show()
```

```python
# 可视化数值属性与目标变量'Churn'之间的关系
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Churn', y=feature, data=df)
    plt.title(f'{feature}与Churn之间的关系')
    plt.show()
```
